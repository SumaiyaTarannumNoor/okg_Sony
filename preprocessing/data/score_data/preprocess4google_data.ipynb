```python
import pandas as pd

# Define the file paths
#file_path1 = '/home/ubuntu/reflexion/OKG/OKG/preprocessing/data/score_data/PO_general.csv'
file_path1 = '/home/ubuntu/reflexion/New_LLM_Agent_4_Ad_Keyword_and_Text/keyword/bitdrive_management_cloud_alltime.csv'
#file_path2 = '/home/ubuntu/reflexion/New_LLM_Agent_4_Ad_Keyword_and_Text/keyword/nnc_old_brand_alltime.csv'
#file_path2 = '/home/ubuntu/reflexion/OKG/OKG/preprocessing/data/score_data/PO_brand.csv'
file_path2 = '/home/ubuntu/reflexion/New_LLM_Agent_4_Ad_Keyword_and_Text/keyword/bitdrive_aws_alltime.csv'
# Read the CSV files
# You might need to adjust the delimiter based on the actual file format
# '\s+' uses regular expression to split on one or more spaces
df1 = pd.read_csv(file_path1, delimiter='\t', skiprows=2, encoding='utf-16')
df2 = pd.read_csv(file_path2, delimiter='\t', skiprows=2, encoding='utf-16')
if len(df1) > 2:
    df1 = df1[:-2]
if len(df2) > 2:
    df2 = df2[:-2]
# Append df2 to df1

# Identify common columns
common_columns = df1.columns.intersection(df2.columns)

# Reindex both dataframes to these common columns
df1_aligned = df1[common_columns]
df2_aligned = df2[common_columns]

result = pd.concat([df1_aligned, df2_aligned], ignore_index=True)

# Display the result
print(result.shape)
print(df1.shape)
print(df2.shape)
```

```python
result
```

```python
# remove several columns by name
result = result.drop(['Keyword status', 'Match type', 'Status', 'Status reasons', 'Currency code'], axis=1)
# Remove triple quotes and square brackets
# Convert the 'Search_Term' column to string
result['Keyword'] = result['Keyword'].astype(str)
result['Keyword'] = result['Keyword'].apply(lambda x: x.replace('"', ''))
result['Keyword'] = result['Keyword'].str.replace('[\[\]]', '', regex=True)
result = result.drop_duplicates(subset='Keyword', keep='first')
# Save to CSV with a tab as the delimiter
result.to_csv('/home/ubuntu/reflexion/OKG/OKG/preprocessing/data/score_data/sony_bitdrive.csv', sep='\t', index=False)
```
